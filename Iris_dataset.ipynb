```python
## Divideed the data into features and labels.


X = data.iloc[:,1:-1].values

Y = data.iloc[:,-1].values
```

```python
submission = pd.DataFrame()
submission['Id'] = data.Id
```

```python
test_data_X = test_data.iloc[:,1:-1].values
test_data_X
```

```python
## Splitting the dataset nto training and test data

from sklearn.model_selection import train_test_split
x_train,x_text,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state = 0)
```

```python
# Scalling The Data 
data.describe()
```

```python
## Scaling the data

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_text)
test_data_X = sc.transform(test_data_X) ## test dataset
```

```python
submission = pd.DataFrame()
submission['ID'] = test_data.ID
```

```python
print('Data type is ',type(x_train))

x_train[:10]
```

# Creating and Training the logistic regression model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(multi_class = 'ovr',random_state = 0)
classifier.fit(x_train,y_train)
```

# Predicting results

```python
from sklearn.metrics import confusion_matrix, accuracy_score
predictions = classifier.predict(x_test)
cm = confusion_matrix(predictions, y_test)
print(cm)
accuracy_score(predictions, y_test)
```
